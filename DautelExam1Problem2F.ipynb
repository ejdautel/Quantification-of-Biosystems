```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
import statsmodels.formula.api as sm
%matplotlib inline
import scipy.stats as stats
from polyfit import polyfit
```

c) The results in plot Leaf Length Over Time show a non-linear relationship between leaf length and time. 

```python

time = [1,8,12,16,20,24,28,32,36,40,44] #dataset
LeafLength = [0.25,1.1,1.75,2.5,3.25,4.5,5.5,5.85,5.95,5.95,5.75]

plt.plot(time,LeafLength, 'bo') #plot dataset
plt.title('Leaf Length Over Time')
plt.ylabel('Length in cm')
plt.xlabel('Time in days')
plt.show()
```

```python
fitData = np.polyfit(time,LeafLength,6) #6th degree polynomial function to fit the leaf length dataset
fitData

#Polynomial fit (6th degree)
p = np.poly1d(fitData)
print(p)

xp= np.linspace(1,45,100) #starting at the first datapoint, fit the curve to a time step of 45
plt.plot(time,LeafLength,'.',xp,p(xp),'-') #plot curve fit over dataset curve
plt.title('Leaf Length Over Time 6th Degree Polynomial Curve Fit')
plt.ylabel('Length in cm')
plt.xlabel('Time in days')
plt.show()
```

```python
#Plot of residuals
yfit3=np.polyval(fitData,time)
plt.plot(time,yfit3-LeafLength,'rx',label='var')
fitData = np.polyfit(time,LeafLength,6)
plt.title('Plot of Residuals')
fitData
 #No pattern in residuals represents good fit
```
