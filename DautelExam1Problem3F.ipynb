a) A simple reaction is two reactants forming a product such that 

A + B = P  	
This results in the differential equations below.
d[P]/dt = k[A][B] and d[A]/dt = d[B]/dt = -k[A][B] 

If reactant A >>> than B it will follow first order kinetics and a straight line will be observed when ln[B] is plotted over time.  Where the concentrations of A and B are unequal but not extremely different both ln[B] and 1/[B] will be curved. Write a program to simulate the reactions in two scenarios below using Euler integration with a dt = 0.1.

Scenario 1:  [A0] = 100 and [B0] = 100 with k = 0.0003
Scenario 2: [A0] = 1000 and [B0] = 10 with k = 0.0003
As the output for each scenario, obtain graphs of 1/[B] and ln[B] plotted against time to determine whether the data sets produce first or second order kinetics.  Assume arbitrary units of concentration and time.  


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

b) When the concentration of A is much larger than B, a first order kinetic reaction is assumed to be observed. Since the kinetic factor affecting both scenario A and scenario B is the same, 0.0003, the concentrations of each reactant will be the only factors determining a first or second order kinetic reaction. 

```python
dt=0.01 #define constants and time parameters
A0=100
B0=100
k=0.0003
tf=100
t0=0
n=1001
dt=(tf-t0)/(n-1)
t=np.linspace(t0,tf,n) 

A = np.zeros([n]) #create empty array of variables 
B = np.zeros([n])
P = np.zeros([n])
```

```python
# First Scenario

A[0]=100 #initial values set 
B[0]=100
P[0]=0
k=0.0003 #kinetic constant set

for i in range(1,n): #euler method of integration 
    A[i]=dt*-k*(A[i-1]*B[i-1])+A[i-1]
    B[i]=dt*-k*(A[i-1]*B[i-1])+B[i-1]

plt.plot(t,np.log(B)) 
plt.title('First Scenario Plot of ln[B]')
plt.ylabel('ln[B]')
plt.xlabel('Time')
plt.show()

plt.plot(t,1/B)
plt.title('First Scenario Plot of 1/[B]')
plt.ylabel('1/[B]')
plt.xlabel('Time')
plt.show()
```
